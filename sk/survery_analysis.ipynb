```python
import pandas as pd 

df = pd.read_csv("survey_cleaned.csv")
```

```python
def get_mode(series):
    return series.mode().iloc[0] if not series.mode().empty else None

mode_df = df.groupby('AudioID')[['Emotion', 'Intensity', 'Gender']].agg(get_mode).reset_index()

count_series = []

for col in ['Emotion', 'Intensity', 'Gender']:
    merged = df.merge(mode_df[['AudioID', col]], on='AudioID', suffixes=('', '_mode'))
    count_col = merged.groupby('AudioID').apply(
        lambda x: (x[col] == x[f'{col}_mode']).sum()
    ).rename(f'{col}_count')
    count_series.append(count_col)

counts_df = pd.concat(count_series, axis=1).reset_index()

counts_df = pd.concat(count_series, axis=1).reset_index()
result_df = mode_df.merge(counts_df, on='AudioID')
```

```python
ids = [i for i in range(1,6)]
emotions =["Angry",	"Calm",	"Disgust","Fearful","Happy","Sad","Surprised"]
intensity = ["Normal","Strong"]
Gender = ["Male","Female"]

count = {}
for id in ids:
    count[id] = {}
    for emotion in emotions:
        count[id][emotion] = {
            "Normal":{"Male":0,"Female":0}, 
            "Strong":{"Male":0,"Female":0}
            }

with open("survey_cleaned.csv") as f:
    data = f.read().splitlines()
    rows = data[1:]
    for row in rows:
        values = row.split(",")
        id = int(values[1])
        e = values[2]
        i = values[3]
        g = values[4]
        count[id][e][i][g] += 1
```
